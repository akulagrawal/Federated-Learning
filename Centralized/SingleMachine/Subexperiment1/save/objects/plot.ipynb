```python
import pickle
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
for (cr,e) in [(10,50),(20,25),(50,10),(100,10)]:
    plt.figure()
    for c in [25, 15, 10, 5]:
        file_name = "mnist_cnn_" + str(cr) + "_C[" + str(c/100) + "]_iid[0]_E[" + str(e) + "]_B[10].pkl"
        csv_name = "_C[" + str(c/100) + "]_CR[" + str(cr) + "]_E[" + str(e) + "].csv"
        try:
            with open(file_name, 'rb') as f:
                loss, acc = pickle.load(f)
            
            if c == 5:
                col='r'
            if c == 10:
                col='b'
            if(c == 15):
                col='y'
            if(c==25):
                col='k'
            plt.plot(range(1,len(acc)+1), acc, color=col, label=str(c/100))
            plt.xticks(np.arange(0, cr+1, cr/10))
            plt.yticks(np.arange(0, 1.1, 0.1))
        except:
            pass
    plt.text(4*cr/5+cr/30, 0.5, "Epochs = " + str(e), fontsize=10)
    plt.text(2*cr/3+cr/14, 0.4, "Total clients = 100", fontsize=10)
    plt.text(3*cr/4, 0.3, "Fraction of clients", fontsize=10)
    plt.ylabel('Average Training Accuracy')
    plt.xlabel('Communication Rounds')
    plt.legend()
    #plt.show()
    plt.savefig(str(cr)+'_'+str(e)+'_acc.png')
```
