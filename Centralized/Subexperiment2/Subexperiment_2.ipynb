```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.title('Communication Rounds vs Fraction of Clients')
for mean_thresh in [0.97]:
    for std_thresh in [0.03]:
        crs = []
        c = [0.05, 0.1, 0.15, 0.25]
        for C in c:
            filename = 'acc_C['+str(C)+']_CR[100]_E[10].csv'
            try:
                df = pd.read_csv(filename, header = None)
            except:
                break
            cr = 0
            for i in range(len(df)-9):
                if (df.rolling(window=10).mean()[0][9+i] > mean_thresh) and (df.rolling(window=10).std()[0][9+i] < std_thresh):
                    print(filename + ": " + str(i+10))
                    crs.append(i+10)
                    break
                else:
                    pass
        plt.plot(c, crs, linestyle='dashed', marker='o', label='mean_lb='+str(mean_thresh)+' std_ub='+str(std_thresh))
        plt.ylabel('Communication Rounds')
        plt.xlabel('Fraction of Clients')
        plt.xticks([0.05, 0.1, 0.15, 0.25])
plt.legend()
#plt.show()
plt.savefig('F_CL.png')
```
