```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
for i in range(1,5):
    plt.figure()
    acc = pd.read_csv("acc_"+str(i)+".csv", header=None)
    plt.plot(range(1,len(acc)+1), acc)
    plt.ylabel('Accuracy')
    plt.xlabel('Agent Rounds')
    #plt.show()
    plt.savefig('acc_'+str(i)+'.png')
```

```python
for i in range(1,5):
    plt.figure()
    loss = pd.read_csv("loss_"+str(i)+".csv", header=None)
    plt.plot(range(1,len(loss)+1), loss)
    plt.ylabel('Loss')
    plt.xlabel('Agent Rounds')
    #plt.show()
    plt.savefig('loss_'+str(i)+'.png')
```
