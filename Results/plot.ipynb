```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.figure()
for n_agents in range(1,4):
    with open("results_"+str(n_agents)+".pkl", "rb") as f:
        train_loss, train_acc, test_acc, test_loss, clients, agents, times = pickle.load(f)
    timet = []
    avg = []
    for i in range(len(test_acc)):
        sum = 0.0
        if(times[i] >= 300000.0):
            break
        if i+19 < len(test_acc):
            for j in range(20):
                sum = sum + test_acc[i+j]
            avg.append(sum*100.0/20.0)
            timet.append(times[i])
            i=i+20
    plt.plot(timet, avg, label = "n_agents = " + str(n_agents))
plt.xlabel("Time (in sec)")
plt.ylabel("% Accuracy, $\\xi$")
plt.yticks(np.arange(0, 100.01, 10))
plt.xticks(np.arange(0, 300050, 50000))
plt.ylim(ymin=0)
plt.xlim(xmin=0)
#plt.text(200000, 37, "FD Version", fontsize=10)
#plt.text(200000, 31, "$\mathregular{F_{Cl}}$ = 20%", fontsize=10)
plt.legend(loc = (0.65, 0.05))
plt.savefig("123_FD.png")
```
